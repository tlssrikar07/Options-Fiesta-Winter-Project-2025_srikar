```python
import numpy as np
import matplotlib.pyplot as plt
# input mehtods
K = float(input("Strike Price: "))
P = float(input("Premium: "))
xmin, xmax = map(float, input("range of spot price(min max): ").split())
option = input("Option Type (C/P/B): ").upper() # C=Call, P=Put, B=Both call and put
```

```python
# call plot
if option == "C":
    fig, ax = plt.subplots(2, 1)
    # creating spot price
    x = np.linspace(xmin, xmax, 100)
    # calculating profit/loss
    long_call = np.maximum(0, x - K) - P
    short_call = P - np.maximum(x - K, 0)
    #ploting the graphs
    ax[0].plot(x, long_call, label="Long Call", color='green')
    ax[0].set_title("Buying Call Option")
    ax[0].set_xlabel("Spot Price")
    ax[0].set_ylabel("Profit / Loss")
    ax[0].legend()
    ax[1].plot(x, short_call, label="Short Call", color='red')
    ax[1].set_title("Selling Call Option")
    ax[1].set_xlabel("Spot Price")
    ax[1].set_ylabel("Profit / Loss")
    ax[1].legend()

    plt.tight_layout()
    plt.show()
# put plot
if option == "P":
    fig, ax = plt.subplots(2, 1)
    # creating spot price
    x = np.linspace(xmin, xmax, 100)
    # calculating profit/loss
    long_put = np.maximum(0, K - x) - P
    short_put = P - np.maximum(K - x, 0)
    #ploting the graphs
    ax[0].plot(x, long_put, label="Long Put", color='green')
    ax[0].set_title("Buying Put Option")
    ax[0].set_xlabel("Spot Price")
    ax[0].set_ylabel("Profit / Loss")
    ax[0].legend()
    ax[1].plot(x, short_put, label="Short Put", color='red')
    ax[1].set_title("Selling Put Option")
    ax[1].set_xlabel("Spot Price")
    ax[1].set_ylabel("Profit / Loss")
    ax[1].legend()

    plt.tight_layout()
    plt.show()
# plot for both call and put
if option == "B":
    fig, ax = plt.subplots(2, 2)
    # creating spot price
    x = np.linspace(xmin, xmax, 100)
    # calculating profit/loss
    long_call = np.maximum(0, x - K) - P
    short_call = P - np.maximum(x - K, 0)
    long_put = np.maximum(0, K - x) - P
    short_put = P - np.maximum(K - x, 0)
    #ploting the graphs
    ax[0 , 0].plot(x, long_call, label="Long Call", color='green')
    ax[0 , 0].set_title("Buying Call Option")
    ax[0 , 0].set_xlabel("Spot Price")
    ax[0 , 0].set_ylabel("Profit / Loss")
    ax[0,0].legend()
    ax[1,0].plot(x, short_call, label="Short Call", color='red')
    ax[1,0].set_title("Selling Call Option")
    ax[1,0].set_xlabel("Spot Price")
    ax[1,0].set_ylabel("Profit / Loss")
    ax[1,0].legend()

    long_put = np.maximum(0, K - x) - P
    short_put = P - np.maximum(K - x, 0)
    ax[0,1].plot(x, long_put, label="Long Put", color='green')
    ax[0,1].set_title("Buying Put Option")
    ax[0,1].set_xlabel("Spot Price")
    ax[0,1].set_ylabel("Profit / Loss")
    ax[0,1].legend()
    ax[1,1].plot(x, short_put, label="Short Put", color='red')
    ax[1,1].set_title("Selling Put Option")
    ax[1,1].set_xlabel("Spot Price")
    ax[1,1].set_ylabel("Profit / Loss")
    ax[1,1].legend()
    plt.tight_layout()
    plt.show()
```
