```python
#question 1
arr = [12, 5, 8, 19, 3, 15]

max = arr[0]
min = arr[0]

for num in arr:
    if num > max:
        max = num
    if num < min:
        min = num

print("Maximum:", max)
print("Minimum:", min)
```

```python
#question 2
string = "radar"
l = 0;
r = len(string)-1
flag = True

while (l < r):
    if (string[l] != string[r]):
        flag = False
        break
    l += 1
    r -= 1

if flag:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#Question 3
arr = [1, 5, 3, 9, 12]
x = 4
arr.sort()

i=0
j=1

while (i < len(arr) and j < len(arr)):
  if (abs(arr[j] - arr[i]) == x):
    print(arr[i], arr[j])
    break
  elif (abs(arr[j] - arr[i]) < x):
    j += 1
  else:
    i += 1
    if (i == j):
      j += 1
```
